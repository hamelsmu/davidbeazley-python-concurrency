```python
class Foo:
    def __init__(self): self.n=10
    def __iter__(self): return self
    def __next__(self): 
        self.n -=1
        if self.n <= 0: raise StopIteration
        return self.n
```

```python
f = Foo()
```

```python
def bar():
    while True:
        f = Foo() 
        yield from f
```
